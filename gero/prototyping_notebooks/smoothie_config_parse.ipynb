```python
def _deserialize_smoothie_config() -> dict[str, str]:
    _myvars = {}
    (counter_lines, counter_skipped_lines) = 0, 0

    config_file_path = "../sample_config_files/config.txt"

    with open(config_file_path) as myfile:
        for line in myfile:
            counter_lines += 1

            # split code and comments for each of the line
            _code, _comment = line.partition("#")[::2]

            if _code.strip() == "":
                counter_skipped_lines += 1
                continue

            # print(_code)
            name, var = _code.partition(" ")[::2]

            # print(f"{name}: {var}")
            myvars[name.strip()] = str(var.strip())
        
        debug(f"line skipped: {skipped_line_count} out of {line_count}")

        return myvars

vars = _deserialize_smoothie_config()
vars
```
